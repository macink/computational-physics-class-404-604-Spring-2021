Write a program to plot

f(x)=\pi/2+2*(sin(x)+sin(3x)/3+sin(5x)/5+....sin(nx)/n)
\end{equation}
from -$\pi$ to $\pi$. Choose n=11, 21 and 31 and plot three curves in one figure. Please use numpy library and matplotlib.

### First I'm going to plot the function with every number up to n=11 to see what the plot is supposed to look like

```python
from pylab import *
import numpy as np

#Use lambda function to create my f(x) expanded to n=11
f=lambda x: (pi/2 + 2*sin(x) + (2/3)*sin(3*x) + (2/5)*sin(5*x) +
             (2/7)*sin(7*x) + (2/9)*sin(9*x) + (2/11)*sin(11*x))

#set up an array of x-values
x=np.arange(-pi,pi,0.1)

#plot the figure
plt.plot(x,f(x))
plt.show()
print('for n=11 the plot should look like this')
```

### Now that I know how the plot should look, I will write a program that can compute this function, f(x), for any value of n

Since the function has a pattern when we use only odd values for n, I will define a function with a summation that looks like

$f(x) = \frac{\pi}{2} + \Sigma_{n=1} (\frac{2}{2n-1})sin((2n-1)x)$.

```python
def f(x,n):
    '''
    A function to compute the required function defined above
    Input: x-values and number of iterations
    Output: the summation
    '''
    values=np.arange(1,n,1)   #array to store the number of iterations
    return pi/2 + sum(2/(2*values-1) * sin((2*values-1)*x)) 
```

Now that we have defined a function, we can loop through our desired x-values and append them to an array for plotting. In the loop we will call our function, f(x) and choose our value for n. From there we can call the plot and show them all together.

```python
#Empty arrays to store the x and y-values for each plot
y1=[]
y2=[]
y3=[]
X=[]   #Since all x-values are the same we only need one X array

#Set the for loops to iterate through all x-values from -pi to pi
for x in np.arange(-pi,pi,0.1):
    y1.append(f(x,11))   #call f(x) for n=11
    X.append(x)
plt.plot(X,y1, label='n=11')
    
for x in np.arange(-pi,pi,0.1):
    y2.append(f(x,21))   #call f(x) for n=21
plt.plot(X,y2, label='n=21')

for x in np.arange(-pi,pi,0.1):
    y3.append(f(x,31))   #call f(x) for n=31
plt.plot(X,y3, label='n=31')

#Set up the plot
xlabel('x', fontsize=14)
ylabel('f(x)', fontsize=14)
title('f(x) vs x', fontsize=16)
legend()
plt.show()
```
